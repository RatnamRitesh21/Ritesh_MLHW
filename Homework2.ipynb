```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculating Gradient Descent values for the loss function with the help of linear equation (y = mx+c)
def LinearModel_Gradient(Learning_Rate, epochs, X,Y):

  m = 0 #slope
  c = 0
  L = Learning_Rate
  y_cap = 0
  n = len(X)

  for i in range(epochs):
    y_cap = m*X + c  #y_cap is the predicted Y vector

    #Calculate the partial derivate(PD) of loss function L=(1/n)*sum((Y-y_cap)**2) wrt m,c
    # To peroform above calculation, we first calculate ∂L/∂(y_cap), ∂(y_cap)/∂m, as L is the function of y_cap and y_cap is the function of 'm'.
    # Then the product of these two derivatives gives us ∂L/∂m = PD_m
    # Similarly we calculate ∂L/∂c = PD_c
    PD_m = (-2/n)*sum(X*(Y-y_cap))
    PD_c = (-2/n)*sum((Y-y_cap))

    # Update parameters using the gradients
    # For each parameter m,c, adjust it in the direction that minimizes the loss
    # by subtracting the product of the learning rate and its respective gradient.
    m = m-(L*PD_m)
    c = c-(L*PD_c)

  y_cap = m*X + c
  return y_cap

# This is the start of our program
# Reading the data from given file
data= pd.read_csv("https://raw.githubusercontent.com/RatnamRitesh21/Ritesh_MLHW/main/HW2_linear_data.csv",header=None)

# Allocating the respective values to two vectors X,Y
Vector_X = data.iloc[:,0]
Vector_Y = data.iloc[:,1]
plt.scatter(Vector_X,Vector_Y,s=15)

# Calculating final predicted vector
predicted_Value = LinearModel_Gradient(0.0001,1000,Vector_X,Vector_Y)

# plotting the predicted Y values with respect to X, having minimum loss function
plt.plot(Vector_X,predicted_Value,color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with Gradient Descent')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculating Gradient Descent values for the loss function with the help of cubic equation (y = a*X**3 + b*X**2+c*X+d)
def Cubic_Gradient(Learning_Rate, epochs, X,Y):
  a,b,c,d = 0,0,0,0
  L = Learning_Rate
  y_cap = 0
  n = len(X)

  for i in range(epochs):
    y_cap = a*X**3 + b*X**2+c*X+d  #y_cap is the predicted Y vector

    # Calculate the partial derivate(PD) of MSE loss function L=(1/n)*sum((Y-y_cap)**2) wrt a,b,c,d
    # To peroform above calculation, we first calculate dL/d(y_cap), d(y_cap)/da, as L is the function of y_cap and y_cap is the function of 'a'.
    # Then the product of these two derivatives gives us dL/da = PD_a
    # Similarly we calculate other partial derivatives (PD_b, PD_c, PD_d) wrt b,c,d.
    PD_a = (-2/n)*sum((Y-y_cap)*X**3)
    PD_b = (-2/n)*sum((Y-y_cap)*X**2)
    PD_c = (-2/n)*sum((Y-y_cap)*X)
    PD_d = (-2/n)*sum((Y-y_cap))

    # Update parameters using the gradients
    # For each parameter a, b, c, and d, adjust it in the direction that minimizes the loss
    # by subtracting the product of the learning rate and its respective gradient.
    a = a-(L*PD_a)
    b = b-(L*PD_b)
    c = c-(L*PD_c)
    d = d-(L*PD_d)

  y_cap = a*X**3 + b*X**2+c*X+d
  return y_cap

# Start of the program
# Reading the data from given file using read_csv function from pandas
data= pd.read_csv("https://raw.githubusercontent.com/RatnamRitesh21/Ritesh_MLHW/main/HW2_nonlinear_data.csv",header=0)

#Assigning the respective values to two vectors X,Y
Vector_X = data.iloc[:,0]
Vector_Y = data.iloc[:,1]
plt.scatter(Vector_X,Vector_Y,s=15)

# Calculating final predicted vector
predicted_Value = Cubic_Gradient(0.000001,10000,Vector_X,Vector_Y)

#plotting the predicted Y values with respect to X, having minimum loss function
plt.scatter(Vector_X,predicted_Value,s=15,color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-Linear Regression with Gradient Descent')
plt.show()
```
